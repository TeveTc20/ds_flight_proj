```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style('darkgrid')
```

```python
df = pd.read_csv('Flights_data_cleaned.csv')
```

```python
df.head()
```

```python
#We can drop Source,Destination and Year columns as they all similar to eachother
df.drop(['Source','Destination','Year'],axis = 1,inplace=True)
```

```python
df.describe().T
```

```python
#Histogram to see how to prices spread
plt.figure(figsize = (14,4))
df.Price.hist(bins = 20)
plt.show()
```

```python
#Histogram to see how to prices spread
plt.figure(figsize = (14,10))
df.Price.plot(kind = 'box')
plt.show()
```

```python
##Checking how much the price depend on total stops of the flight
plt.figure(figsize = (14,4))
sns.lineplot(x = 'Total stops' , y = 'Price' , data = df)
plt.show()
```

```python
plt.figure(figsize = (14,4))
sns.barplot(x = 'Duration' , y = 'Price' , data = df)
plt.show()
```

```python
plt.figure(figsize = (14,4))
sns.barplot(x = 'Airline' , y = 'Price' , data = df,color = 'seagreen')
plt.xticks(rotation=80)
plt.show()
```

```python
plt.figure(figsize = (14,4))
df.Duration.hist(bins = 20)
plt.show()
```

```python
plt.figure(figsize = (14,4))
df.Month.hist(bins = 50)
plt.show()
```

```python
plt.figure(figsize = (14,4))
sns.barplot(x = 'Month' , y = 'Price' , data = df)
plt.show()
```

```python
plt.figure(figsize = (14,4))
sns.barplot(x = 'Duration' , y = 'Total stops',data = df,color = 'orange')
plt.show()
```

```python
plt.figure(figsize = (14,4))
plt.subplot(1,2,1)
sns.lineplot(x = 'Departure time' , y = 'Price' , data = df)
plt.subplot(1,2,2)
sns.lineplot(x = 'Arrival time' , y = 'Price' , data = df)
plt.show()
```

```python
sns.heatmap(df.drop('Airline',axis = 1).corr(),annot=True)
```

# In conclusion

### We can see that the flight ticket prices from TLV to BKK are in average 500 Euros.
